```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('등급별예측데이터.csv',encoding="cp949")
df.head(5)
```

```python
df.columns
```

```python
df.dtypes
```

```python
# 숫자형 변수
num = df.select_dtypes(include=['int64', 'float64'])

# 범주형 변수
cat = df.select_dtypes(include=['object', 'category'])
```

```python
cat = cat.drop('창업성공등급', axis = 1) # 종속 변수 제외
```

```python
num
```

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
num_scaled = scaler.fit_transform(num)
num_df_scaled=pd.DataFrame(data=num_scaled, columns=num.columns)
num_df_scaled
```

```python
cat_dummies=pd.get_dummies(cat).astype(np.float32)
cat_dummies
```

```python
a=df[['창업성공등급']]
```

```python
comp=pd.concat([a, num_df_scaled, cat_dummies],axis=1)
comp
```
